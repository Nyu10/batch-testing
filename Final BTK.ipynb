```python
import Nehemiah_Batch_Testing as BTK #BTK is batch testing kit
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
df = pd.DataFrame([], columns = ['Infection_Rate', 'Sequential_Test', 'Stop_Rule', 'Sequential_Repeats','Threshold', 'Accuracy', 'Accuracy_STD', 'Sensitivity', 'Sensitiviy_STD', 'Specificity','Specificity_STD','PPV', 'PPV_STD',
    'NPV', 'NPV_SD', 'Total_Test_Consumption', 'Test_Test_Consumption_SD', 'Individual_Test_Consumption', 'Individual_Test_Consumption_SD', 'Batch_Test_Consumption','Batch_Test_Consumption_SD'])

Constants = BTK.setupConstants(typeIIError = 0.15, typeIError = 0.01, batchLimit = 32)
for infectionRates in [0.001, 0.01, 0.03, 0.05, 0.10]:
    populationArrayIterations = [BTK.setupPopulation(100000, infectionRates) for iterations in range(100)]
    batchSize = BTK.optimalBatchSize(0.01, Constants)
    kwargs = {'stopRule': 3, 'p': infectionRates, 'batchSize': batchSize,
                         'Constants': Constants, 'seqRepeat': 3, 'threshold': 0.3, 'sequential': True}
    dataFrame = BTK.resultsToDataFrame(populationArrayIterations, **kwargs)
    dfMean = dataFrame.mean()
    dfSTD = dataFrame.std()
    temp = [kwargs['p'], kwargs['sequential'], kwargs['stopRule'], kwargs['seqRepeat'], kwargs['threshold'],dfMean['accuracy'], dfSTD['accuracy'], dfMean['sensitivity'], dfSTD['sensitivity'], dfMean['specificity'], dfSTD['specificity'], dfMean['PPV'], dfSTD['PPV'], dfMean['NPV'], dfSTD['NPV'], dfMean['totalTestConsumption'], dfSTD['totalTestConsumption'], dfMean['individualTestConsumption'], dfSTD['individualTestConsumption'], dfMean['batchTestConsumption'], dfSTD['batchTestConsumption']]
    newDF = pd.DataFrame(temp)
    newDF = newDF.T
    newDF.columns = ['Infection_Rate', 'Sequential_Test', 'Stop_Rule', 'Sequential_Repeats','Threshold', 'Accuracy', 'Accuracy_STD', 'Sensitivity', 'Sensitiviy_STD', 'Specificity','Specificity_STD','PPV', 'PPV_STD',
    'NPV', 'NPV_SD', 'Total_Test_Consumption', 'Test_Test_Consumption_SD', 'Individual_Test_Consumption', 'Individual_Test_Consumption_SD', 'Batch_Test_Consumption','Batch_Test_Consumption_SD']
    df = pd.concat([df, newDF])
    print(infectionRates)
print(df)
```
